## EDA

```python
sns.countplot(y=df.Gender)
```

```python
cat_col = []
for i in df:
    if len(df[i].unique())<=20:
        cat_col.append(i)
```

```python
cat_col
```

```python
df_cat = df[cat_col]
df_cat
```

```python
cont_col = []
for i in df:
    if len(df[i].unique())>=20:
        cont_col.append(i)
```

```python
cont_col
```

```python
df_cont = df[cont_col]
df_cont
```

```python
plt.figure(figsize=(10,30))
plotnumber = 1
for i in df_cat:
    plt.subplot(14,2,plotnumber)
    sns.countplot(x=df_cat[i])
    plt.xticks(rotation =20)
    plotnumber = plotnumber+1
plt.tight_layout()
```

## Insights

```python
plt.figure(figsize=(10,30))
plotnumber = 1
for i in df_cont:
    plt.subplot(5,3,plotnumber)
    sns.histplot(x=df_cont[i], kde = True)
    plotnumber = plotnumber+1
plt.tight_layout()
```

```python
plt.figure(figsize=(10,30))
plotnumber = 1
for i in df_cat:
    plt.subplot(10,3,plotnumber)
    sns.countplot(x=df_cat[i], hue = df.Gender)
    plt.xticks(rotation = 30)
    plotnumber = plotnumber+1
plt.tight_layout()
```
